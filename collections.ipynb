# 1. Data Structures (Built-in Collections)

These store and organize data.

| Type | Concept |
|-----|--------|
| List | Ordered mutable collection |
| Tuple | Ordered immutable collection |
| Dictionary | Key–value mapping |

These belong to:

**Python Built-in Data Structures / Collections**

Example:

```python
numbers = [1,2,3]        # list
point = (10,20)          # tuple
person = {"name":"Nipun"} # dictionary

````python
# I/P [1,2,3,4,5,6] -> O/P [2,4,6]

numbers = [1,2,3,4,5,6]
result = list(filter(lambda x:x%2==0,numbers))
print(result)
````

````python
# I/P [1,2,3,4,5,6] -> O/P [1, 4, 9, 16, 25]

numbers = [1,2,3,4,5]
result = list(map(lambda x: x**2, numbers))
print(result)
````

````python
# ["apple","dog","banana","cat"] -> ["apple","banana"] (words longer than 4 letters)
words = ["apple", "dog", "banana", "cat"]

words_less_than_4_letters = list(filter(lambda x : len(x)>4, words))

print(words_less_than_4_letters)
````

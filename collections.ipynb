# 1. Data Structures (Built-in Collections)

These store and organize data.

| Type | Concept |
|-----|--------|
| List | Ordered, mutable collection |
| Tuple | Ordered, immutable collection |
| Dictionary | Key–value mapping |
| Set | Unordered collection of unique elements |
| Frozen Set | Immutable version of a set |

These belong to:

**Python Built-in Data Structures / Collections**

Example:

```python
numbers = [1, 2, 3]              # List
point = (10, 20)                 # Tuple
person = {"name": "Nipun"}       # Dictionary
unique_ids = {101, 102, 103}     # Set
fixed_set = frozenset([1, 2, 3]) # Frozen Set
```

---

# 2. Quick Characteristics

| Structure | Ordered | Mutable | Allows Duplicates |
|---|---|---|---|
| List | Yes | Yes | Yes |
| Tuple | Yes | No | Yes |
| Set | No | Yes | No |
| Frozen Set | No | No | No |
| Dictionary | Yes* | Yes | Keys: No |

\* Dictionaries preserve insertion order in **Python 3.7+**.

---

# 3. When to Use What (Intuition)

| Use Case | Best Structure |
|---|---|
| Sequence of items you will modify | List |
| Fixed data that shouldn't change | Tuple |
| Fast lookup by key | Dictionary |
| Remove duplicates / membership check | Set |
| Immutable unique collection | Frozen Set |

Example:

```python
# fast lookup
user_age = {"Nipun": 35}

# unique elements
visited_pages = {"home", "products", "cart"}

# fixed coordinates
location = (28.61, 77.23)
```

# 5. Advanced Collections (Standard Library)

Python also provides additional data structures in the `collections` module.

| Type | Purpose |
|---|---|
| deque | Fast append/pop from both ends |
| Counter | Counting elements |
| defaultdict | Dictionary with default values |
| namedtuple | Tuple with named fields |
| OrderedDict | Ordered dictionary (mainly used before Python 3.7) |

Example:

```python
from collections import Counter

Counter(["apple", "banana", "apple"])
```

Output:

```
Counter({'apple': 2, 'banana': 1})
```

# 6. Core Python Collections (Summary)

The five fundamental built-in collections are:

Most Python data structures and libraries build on top of these.
List comprehension:
```python
[expression for item in iterable if condition]
```

````python
# I/P [1,2,3,4,5,6] -> O/P [2,4,6]

numbers = [1,2,3,4,5,6]
result = list(filter(lambda x:x%2==0,numbers))
print(result)
````

````python
# I/P [1,2,3,4,5,6] -> O/P [1, 4, 9, 16, 25]

numbers = [1,2,3,4,5]
result = list(map(lambda x: x**2, numbers))
print(result)
````

````python
# ["apple","dog","banana","cat"] -> ["apple","banana"] (words longer than 4 letters)
words = ["apple", "dog", "banana", "cat"]

words_less_than_4_letters = list(filter(lambda x : len(x)>4, words))

print(words_less_than_4_letters)
````

````python
# names = ["Alice","Bob","Charlie"] scores = [85,92,88] -> [("Alice",85),("Bob",92),("Charlie",88)]

Names = ["Alice", "Bob", "Charlie"]
Scores = [85, 92, 88]

result = zip(Names, Scores)

print(*result)
````

````python
"""
keys = ["name","age","city"]
values = ["Nipun",25,"New York"]

result = {
          "name":"Nipun",
          "age":35,
          "city":"Delhi"
        }

"""

keys = ["name", "age", "city"]
values = ["Nipun", 35, "Delhi"]

# result_1 = zip(keys, values)
result_2 = dict(zip(keys, values))


# print(*result_1)
print(result_2)
````

````python
"""
A = [1,2,3]
B = [4,5,6]

[4,10,18]
"""

A = [1,2,3]
B = [4,5,6]

# Intutive way
Z = zip(A,B)
L = []
for a, b in Z:
	L.append(a*b)
print(L)

# Expert way
result = [a*b for a,b in zip(A,B)]
print(result)
````

````python
"""
Fruits = ["apple", "banana", "mango"]

0 apple
1 banana
2 mango

"""

Fruits = ["apple", "banana", "mango"]
for a, b in enumerate(Fruits):
  print(a, b)

B = dict(enumerate(Fruits))
print(B)


````

````python
"""
[-3,-1,0,4,7]

[4, 7]

"""


Num = [-3,-1,0,4,7]

# list(filter(lambda x: cond, items))
print(list(filter(lambda x: x>0, Num)))

# [expression for item in iterable if condition]
print([x for x in Num if x>0])

````

````python
"""
[5,2,8,1]


[1,2,5,8]

"""

# sequence[start : stop : step]
print(Num[::-1])
````

````python
"""
[5,2,8,1]


[(1,5),(8,2)]

"""

NUM = [5,2,8,1]


Result = list(zip(NUM[::-1],NUM))
print(Result[:2])


Result = [(x, y) for x, y in zip(NUM[::-1],NUM)]
print(Result[:2])
````
